```python
df = df.apply(lambda x: x.replace(" ", np.NaN))
```

```python
df['GEOCODE2'].value_counts()
```

```python
m = df['GEOCODE2'].mode()[0]
df['GEOCODE2'] = df['GEOCODE2'].fillna(m)
```

```python
df['GEOCODE2'].value_counts()
```

```python
df.isna().sum()
```

```python
df.isna().sum()/len(df)
```

Since the number of missing value in MSA, ADI, DMA are less and they are codes,I am filling nas with mode.

```python
n = df['MSA'].mode()[0]
df['MSA'] = df['MSA'].fillna(n)
```

```python
o = df['ADI'].mode()[0]
df['ADI'] = df['ADI'].fillna(o)
```

```python
p = df['DMA'].mode()[0]
df['DMA'] = df['DMA'].fillna(p)
```
